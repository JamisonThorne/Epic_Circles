```python
from math import isclose, sqrt
```

```python
def are_circles_tangent(circle1, circle2):
    #determines if two circles are tangent
    # circle = [x, y, r]
    x1, y1, r1 = circle1
    x2, y2, r2 = circle2
    center_dist = sqrt((x1 - x2)**2 + (y1 - y2)**2)
    # Check for external tangency
    if isclose(center_dist, r1 + r2, rel_tol=0.01):
        return True
    # Check for internal tangency
    if isclose(center_dist, abs(r1 - r2), rel_tol=0.01):
        return True
    return False

def find_my_tangent_circle(tangential_circles,kissing_circles):
    for tangential_circle_check in kissing_circles:
        counter = 0
        for tangent_circle in tangential_circles:
            if are_circles_tangent(tangential_circle_check, tangent_circle):
                counter+=1
        if counter == 3:
            tangential_circles.append(tangential_circle_check)
    return tangential_circles
```

```python
tangential_circles = [[0.0, 0.0, 50.0], [25.0, 0.0, 25.0], [-25.0, 0, 25.0]]
kissing_circles = [[0.0, 33.33, 16.67], [-0.0, -33.33, 16.67]]
```

```python
result = find_my_tangent_circle(tangential_circles, kissing_circles)
print(result)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

```python
fig, ax = plt.subplots()

tangential_circles = [[0.0, 0.0, 50.0], [25.0, 0.0, 25.0], [-25.0, 0, 25.0]]
kissing_circles = [[0.0, 33.33, 16.67], [-0.0, -33.33, 16.67]]
for circle in tangential_circles:
    center = (circle[0], circle[1])
    radius = circle[2]
    circle = Circle(center, radius, fill=False)
    ax.add_patch(circle)
    ax.set_aspect(1)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
for circle in kissing_circles:
    center = (circle[0], circle[1])
    radius = circle[2]
    circle = Circle(center, radius, fill=False)
    ax.add_patch(circle)
    ax.set_aspect(1)
    ax.set_xlim(-51, 51)
    ax.set_ylim(-51, 51)
plt.show()
```
